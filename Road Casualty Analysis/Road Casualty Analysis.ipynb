```python
import pandas as pd
df = pd.read_csv("C:/Users/Hing/Downloads/dft-road-casualty-statistics-casualty-2020.csv")
```

```python
df
```

```python
df1 = df.loc[:, ['accident_reference', 'casualty_class', 'sex_of_casualty', 'age_of_casualty', 'age_band_of_casualty', 'casualty_severity','pedestrian_location', 'pedestrian_movement','car_passenger','casualty_type','casualty_home_area_type','casualty_imd_decile']]
df1
```

```python
df2 = df1.loc[( (df["sex_of_casualty"] == 1) | (df["sex_of_casualty"] == 2) ) 
              & (df["pedestrian_location"] != -1)
              & (df["pedestrian_movement"] != -1)
              & (df["car_passenger"] != -1)
              & (df["casualty_home_area_type"] != -1)
              & (df["casualty_imd_decile"] != -1)
              & (df["age_band_of_casualty"] != -1)]
df2
```

```python
dg = pd.read_excel("C:/Users/Hing/Downloads/Road-Safety-Open-Dataset-Data-Guide.xlsx")
dg
```

```python
dg1 = dg.loc[(dg["table"] == 'Casualty') ]
dg1
```

```python
dgsex = dg1.loc[(dg["field name"] == 'sex_of_casualty')]
dgsex = dgsex.loc[:, ['code/format', 'label']]
dgsex = dgsex.rename(columns={'code/format': 'sex_of_casualty', 'label': 'Sex'})
dgsex
```

```python
df2 = df2.merge(dgsex, on='sex_of_casualty', how='left')
df2
```

```python
dgclass = dg1.loc[(dg["field name"] == 'casualty_class')]
dgclass = dgclass.loc[:, ['code/format', 'label']]
dgclass = dgclass.rename(columns={'code/format': 'casualty_class', 'label': 'Class'})
dgclass
```

```python
df2 = df2.merge(dgclass, on='casualty_class', how='left')
df2
```

```python
dgageband = dg1.loc[(dg["field name"] == 'age_band_of_casualty')]
dgageband = dgageband.loc[:, ['code/format', 'label']]
dgageband = dgageband.rename(columns={'code/format': 'age_band_of_casualty', 'label': 'Age_Band'})
dgageband
```

```python
dgageband["Age_Band"].replace({"0 - 5": "00 - 05",
                               "6 - 10": "06 - 10"}, inplace=True)
dgageband
```

```python
df2 = df2.merge(dgageband, on='age_band_of_casualty', how='left')
df2
```

```python
dgseverity = dg1.loc[(dg["field name"] == 'casualty_severity')]
dgseverity = dgseverity.loc[:, ['code/format', 'label']]
dgseverity = dgseverity.rename(columns={'code/format': 'casualty_severity', 'label': 'Severity'})
dgseverity
```

```python
df2 = df2.merge(dgseverity, on='casualty_severity', how='left')
df2
```

```python
dglocation = dg1.loc[(dg["field name"] == 'pedestrian_location')]
dglocation = dglocation.loc[:, ['code/format', 'label']]
dglocation = dglocation.rename(columns={'code/format': 'pedestrian_location', 'label': 'Location'})
dglocation
```

```python
dglocation["Location"].replace({"Crossing on pedestrian crossing facility": "Crossing facility", 
                                "Crossing in zig-zag approach lines": "Zig-zag lines",
                                "In carriageway, crossing elsewhere": "Carriageway",
                                "In centre of carriageway - not on refuge, island or central reservation": "Carriageway",
                                "In carriageway, not crossing": "Carriageway",
                                "Crossing elsewhere within 50m. of pedestrian crossing": "Within 50m of pedestrian crossing",
                                "Crossing in zig-zag exit lines": "Zig-zag lines" }, inplace=True)
dglocation
```

```python
df2 = df2.merge(dglocation, on='pedestrian_location', how='left')
df2
```

```python
dgvehicle = dg1.loc[(dg["field name"] == 'casualty_type')]
dgvehicle = dgvehicle.loc[:, ['code/format', 'label']]
dgvehicle = dgvehicle.rename(columns={'code/format': 'casualty_type', 'label': 'Vehicle'})
dgvehicle
```

```python
dgvehicle["Vehicle"].replace({"Motorcycle 50cc and under rider or passenger": "Motorcycle", 
                                "Motorcycle 125cc and under rider or passenger": "Motorcycle",
                                "Motorcycle over 125cc and up to 500cc rider or  passenger": "Motorcycle",
                                "Motorcycle over 500cc rider or passenger": "Motorcycle",
                                "Goods vehicle (over 3.5t. and under 7.5t.) occupant": "Goods vehicle",
                                "Goods vehicle (7.5 tonnes mgw and over) occupant": "Goods vehicle",
                                "Motorcycle - unknown cc rider or passenger": "Motorcycle",
                                "Van / Goods vehicle (3.5 tonnes mgw or under) occupant": "Goods vehicle",
                              "Tram occupant": "Tram",
                              "Taxi/Private hire car occupant": "Taxi/Private hire",
                              "Minibus (8 - 16 passenger seats) occupant": "Minibus",
                              "Bus or coach occupant (17 or more pass seats)": "Bus or coach",
                              "Agricultural vehicle occupant": "Agricultural vehicle",
                              "Mobility scooter rider": "Mobility scooter",
                              "Electric motorcycle rider or passenger": "Electirc motorcycle",
                              "Other vehicle occupant": "Other vehicle",    
                              "Car occupant": "Car",
                              "Horse rider": "Horse",
                              "Cyclist": "Bicycle",
                                "Goods vehicle (unknown weight) occupant": "Goods vehicle"}, inplace=True)
dgvehicle
```

```python
df2 = df2.merge(dgvehicle, on='casualty_type', how='left')
df2
```

```python
dgimd = dg1.loc[(dg["field name"] == 'casualty_imd_decile')]
dgimd = dgimd.loc[:, ['code/format', 'label']]
dgimd = dgimd.rename(columns={'code/format': 'casualty_imd_decile', 'label': 'IMD_decile'})
dgimd
```

```python
df2 = df2.merge(dgimd, on='casualty_imd_decile', how='left')
df2
```

```python
dgarea = dg1.loc[(dg["field name"] == 'casualty_home_area_type')]
dgarea = dgarea.loc[:, ['code/format', 'label']]
dgarea = dgarea.rename(columns={'code/format': 'casualty_home_area_type', 'label': 'Area_type'})
dgarea
```

```python
df2 = df2.merge(dgarea, on='casualty_home_area_type', how='left')
df2
```

```python
df3 = df2.loc[:, ['accident_reference', 'Class', 'Sex', 'age_of_casualty', 'Age_Band', 'Severity','Location', 'Vehicle','Area_type','IMD_decile']]
df3
```

```python
df3 = df3.rename(columns={'age_of_casualty': 'Age'})
df3
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
color = ['green','yellow','red']
re1 = sns.histplot(data=df3, x="Severity", hue = 'Severity', palette = color).set(title='Road Casualty Severity', ylabel='Number of cases')
```

```python
re2 = sns.histplot(data=df3, x="Class", hue = 'Class').set(title='Road Casualty Person', xlabel='Person', ylabel='Number of cases')
```

```python
df_fatal_d = df3.loc[ (df3["Severity"] == "Fatal") 
                    & (df3["Class"] == "Driver or rider")]
df_fatal_d
```

```python
color = ['skyblue','lightgreen']
re2 = sns.histplot(data=df_fatal_d, x="Sex", hue = 'Sex', palette= color).set(title='Fatal case by Gender', xlabel='Gender', ylabel='Number of cases')
```

```python
c = df_fatal_d.Sex.value_counts(dropna=False)
p = 100 * df_fatal_d.Sex.value_counts(dropna=False, normalize=True)
pd.concat([c,p], axis=1, keys=['counts', '%'])
```

```python
plot = p.plot.pie(title = "Fatal case by Gender by %", autopct = '%1.1f%%', colors = ['skyblue','lightgreen'])
```

```python
c_age = df_fatal_d.Age_Band.value_counts().sort_index()
c_age
```

```python
lines = c_age.plot.line(title = "Age of Fatal case", figsize = (5,5))
lines.set_xlabel("Age Band")
lines.set_ylabel("Number of cases")
```
